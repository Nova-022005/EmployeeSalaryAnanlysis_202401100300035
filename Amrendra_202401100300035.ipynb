IMPORTING THE NECESSARY LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

ADD THE DATASET

```python
# Load the CSV file
df = pd.read_csv('/content/employee_data.csv')
```

```python
print("Data Types:\n", df.dtypes)
```

```python
# Display the first few rows of the dataset
print(df.head())
```

PLOTTING THE GRAPH

```python
# Plot salary distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Salary'], bins=20, kde=True, color='blue')
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
# Scatter plot: Salary vs. Experience
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Experience', y='Salary', data=df, hue='Department', palette='coolwarm', s=100)
plt.title('Salary vs. Experience')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
# Scatter plot: Salary vs. Age
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Salary', data=df, hue='Department', palette='magma', s=100)
plt.title('Salary vs. Age')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
# Scatter plot: Experience vs. Age
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Experience', data=df, hue='Department', palette='Set2', s=100)
plt.title('Experience vs. Age')
plt.xlabel('Age')
plt.ylabel('Years of Experience')
plt.show()
```
